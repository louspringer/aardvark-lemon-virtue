### Few shot learning
using some few shot learning to solve some issues

```python
qns1 = db_chain("How many matches did Egypt played at AFCON 2023? ")
```

```python
qns2 = db_chain("How won AFCON 2023 title and who scored goals? ")
```

```python
qns3 = db_chain("Who won Women's World Cup title  and who scored most goals at the whole champion?")
```

```python
qns4 = db_chain("Depending on the data you have, Who is the best team at Fifa world cup 2022 and who is the worst depending on all champion matches not just final but for all stages including group stages,round 16,quarter final, semi final and final ?")
```

```python
few_shots = [
    {'Question' : "How many matches did Egypt played at AFCON 2023?",
     'SQLQuery' : "SELECT COUNT(*) AS matches_played FROM afcon_2023 WHERE home_team = 'Egypt' OR away_team = 'Egypt'",
     'SQLResult': "Result of the SQL query",
     'Answer' : qns1},
    {'Question': "How won AFCON 2023 title and who scored goals?",
     'SQLQuery':"SELECT home_team, home_score, away_team, away_score FROM afcon_2023 WHERE competition_stage = 'Final'",
     'SQLResult': "Result of the SQL query",
     'Answer': qns2},
    {'Question': "Who won Women's World Cup 2023 and who scored ? " ,
     'SQLQuery' : """SELECT home_team, home_score, away_team, away_score, competition_stage, home_managers,
      away_managers FROM women_wc_2023 WHERE competition='International - Women''s World Cup' AND competition_stage='Final'""",
     'SQLResult': "Result of the SQL query",
     'Answer': qns3} ,
     {'Question' : "Depending on the data you have, Who is the best team at Fifa world cup 2022 and who is the worst ?" ,
      'SQLQuery': "SELECT competition, home_team, home_score, away_team, away_score, competition_stage FROM wc_2022 WHERE competition = 'International - FIFA World Cup' AND season = 2022 ORDER BY home_score DESC LIMIT 5",
      'SQLResult': "Result of the SQL query",
      'Answer' : qns4
      }
]
```
